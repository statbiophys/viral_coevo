```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
f = open('data/params.tsv', 'r')

Nhs = [100000000, 1000000000]
N_Uds = 9
Ud_list, r0_list, D_list = [[], []], [[], []], [[], []]
N_list, ud_list, sigma_list = [[], []], [[], []], [[], []]
z_tmrca = 1.66
speed = 2.6
gamma = 0.2*365

for line in f.readlines():
    if line[0] == '#':
        continue
        
    if float(line.split('\t')[0]) == Nhs[0]:
        i = 0
    elif float(line.split('\t')[0]) == Nhs[1]:
        i = 1
        
    Ud_list[i].append(float(line.split('\t')[1]))
    r0_list[i].append(float(line.split('\t')[2]))
    D_list[i].append(float(line.split('\t')[3]))
    N_list[i].append(float(line.split('\t')[4]))
    ud_list[i].append(float(line.split('\t')[5]))
    sigma_list[i].append(float(line.split('\t')[6]))
    
f.close()

D_list = np.array(D_list)
N_list = np.array(N_list)
ud_list = np.array(ud_list)
sigma_list = np.array(sigma_list)
```

```python
colors = ['#033500', '#40a368']
markers = ['x', '+']
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(3.5,6))

ax1.set_ylabel(r'$r_0$', fontsize=12)
ax2.set_ylabel(r'$\mu_x$', fontsize=12)
ax2.set_xlabel(r'$U_d$', fontsize=12)
for i in range(len(Nhs)):
    ax1.scatter(Ud_list[i], r0_list[i], label=r'$Nh$=%g'%Nhs[i], c=colors[i], marker=markers[i], s=100)
    ax1.plot(Ud_list[i], r0_list[i], c=colors[i])
    ax2.scatter(Ud_list[i], D_list[i]*2, label=r'$Nh$=%g'%Nhs[i], c=colors[i], marker=markers[i], s=100)
    ax2.plot(Ud_list[i], D_list[i]*2, c=colors[i])
    
plt.tight_layout()
ax1.legend()
plt.savefig('plots/data_pars.svg')
```
