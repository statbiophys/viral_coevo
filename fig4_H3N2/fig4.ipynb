```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
f = open('data/params.tsv', 'r')

Nhs = [100000000, 1000000000]
N_Uds = 9
Ud_list, r0_list, D_list = [[], []], [[], []], [[], []]
for line in f.readlines():
    if line[0] == '#':
        continue
        
    if float(line.split('\t')[0]) == Nhs[0]:
        i = 0
    elif float(line.split('\t')[0]) == Nhs[1]:
        i = 1
        
    Ud_list[i].append(float(line.split('\t')[1]))
    r0_list[i].append(float(line.split('\t')[2]))
    D_list[i].append(float(line.split('\t')[3]))
    
f.close()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(6,3))

for i in range(len(Nhs)):
    ax1.scatter(Ud_list[i], r0_list[i], label=r'$Nh$=%g'%Nhs[i])
    ax1.plot(Ud_list[i], r0_list[i])
    ax2.scatter(Ud_list[i], D_list[i], label=r'$Nh$=%g'%Nhs[i])
    ax2.plot(Ud_list[i], D_list[i])
    
plt.tight_layout()
ax1.legend()
plt.savefig('data_pars.svg')
```
