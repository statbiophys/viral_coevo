```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../src/')
import wave_virulence as wv
import utils as ut
import plotter
import numba_functions as nbf
import numba as nb
from copy import deepcopy
import pickle
```

```python
def build_assay_wave(id_, r0, D, beta0, stoch=False, gamma=0.5, cutoff=10000, n_alpha_bins=25, time=500, eps=1e-5, global_cutoff=True,
                     alphas=None, ai0=None):
    dx = ut.lin_from_two_points(np.log(r0), np.log(0.01), 0.001, np.log(1), 0.005)
    dt = ut.dt_from_cfl(0.03, D, dx)

    p = wv.Vwave_virulence_pars(id_, tot_time=time, dt=dt, dx=dx, 
                     n_x_bins=int( ut.lin_from_two_points(r0, 0.01, 3, 1, 50)/dx ), 
                     M=5, r0=r0, D_coef=D, beta=beta0, gamma=gamma,
                     n_alpha_bins=n_alpha_bins, eps=eps, alphas=alphas, alpha0_i=ai0,
                     Nh=10**12, N0=10**10, cutoff=cutoff,
                     t_burn=ut.lin_from_two_points(r0, 0.01, 250, 1, 4000), 
                     t_burn_cutoff=ut.lin_from_two_points(r0, 0.01, 70, 1, 1000), 
                     back_width_fract=ut.lin_from_two_points(np.log(r0), np.log(0.0075), 0.1, np.log(1), 3),
                     traj_step=int(5/dt), check_step=0, traj_after_burn=True, verbose=True, global_cutoff=global_cutoff)
    
    return deepcopy(wv.Vwave_virulence(p, stoch))
```

## r0 scan of ES virulence in the stochastic setting

```python
r0_list = np.logspace(np.log10(0.015), np.log10(1), 15)
D_coef = 1e-5
gamma = 0.5
stoch, cutoff = True, 100000
n_alpha_bins = 100
eps = 2e-6
time = 30000
```

```python
assays = []
alpha_guess=[0.99,0.99,0.98,0.97,0.96,0.95, 0.93, 0.9, 0.85, 0.80, 0.76, 0.71, 0.67, 0.64, 0.6]
alphas = np.linspace(0.3, 1.5, n_alpha_bins)
for i, r0 in enumerate(r0_list):
    a = alpha_guess[i]
    ai0 = int((alphas<a).sum())
    assays.append(build_assay_wave(i, r0, D_coef, 2, stoch, gamma, cutoff, n_alpha_bins, time=time, global_cutoff=False, eps=eps,
                                  alphas=alphas, ai0=ai0))
```

### Importing the different runs and merging them together

```python
f = open('data/virulence_stoch_1_assays.pickle', 'rb')
assays1 = pickle.load(f)
f.close()
f = open('data/virulence_stoch_2_assays.pickle', 'rb')
assays2 = pickle.load(f)
f.close()
f = open('data/virulence_stoch_3_assays.pickle', 'rb')
assays3 = pickle.load(f)
f.close()
f = open('data/virulence_stoch_4_assays.pickle', 'rb')
assays4 = pickle.load(f)
f.close()
assay_list = [assays1, assays2, assays3, assays4]
```

```python
mean_alphas, std_alphas = np.zeros(len(assays1)), np.zeros(len(assays1))
av_time_start = 2000
speeds, f_tips, s_tips = np.zeros(len(assays1)),np.zeros(len(assays1)), np.zeros(len(assays1))

for ass in assay_list:
    for i, w in enumerate(ass):
        av_time_start_i = (w.traj.times < av_time_start).sum()
        speeds[i] += np.mean(w.traj.speed(10)[av_time_start_i:]) / len(assay_list)
        f_tips[i] += np.mean(w.traj.f_tip[av_time_start_i:]) / len(assay_list)
        s_tips[i] += np.mean(w.traj.s_tip[av_time_start_i:]) / len(assay_list)
        mean_alphas[i] += np.mean(w.traj.mean_alpha[av_time_start_i:]) / len(assay_list)
        std_alphas[i] += np.std(w.traj.mean_alpha[av_time_start_i:]) / len(assay_list)
```

```python
f = open('data/ES_virulence_stoch.tsv', 'w')

header = '#r0\talpha_ES\talpha_ES_std\tF_T\ts_T\n'
f.write(header)

for i in range(len(r0_list)):
    f.write(str(r0_list[i]) + '\t')
    f.write(str(mean_alphas[i]) + '\t')
    f.write(str(std_alphas[i]) + '\t')
    f.write(str(f_tips[i]) + '\t')
    f.write(str(s_tips[i]) + '\t')
    
    f.write('\n')
    
f.close()


f = open('data/ES_virulence_stoch_pars.tsv', 'w')

header = '#parameter\tvalue\n'
f.write(header)

f.write('D\t' + str(assays1[0].p.D_coef) + '\n')
f.write('b\t' + str(assays1[0].p.beta) + '\n')
f.write('gamma\t' + str(assays1[0].p.gamma) + '\n')
f.write('Nh\t' + str(assays1[0].p.Nh) + '\n')
f.write('M\t' + str(assays1[0].p.M) + '\n')
f.write('stoch_cutoff\t' + str(assays1[0].p.cutoff) + '\n')

f.close()
```
