```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../src/')
import wave_mut_rate as wD
import utils as ut
import plotter
import numba_functions as nbf
import numba as nb
from copy import deepcopy
from scipy.optimize import brentq
import pickle
```

```python
def build_assay_wave(id_, r0, beta, lambda_tilde, Ds, gamma=1, cutoff=10000, n_D_bins=50, time=10000, eps=0.5e-13, init_n=None, init_nh=None, D0_i=None):
    dx = ut.lin_from_two_points(np.log(r0), np.log(0.05), 0.012, np.log(5), 0.06)
    dt = ut.dt_from_cfl(0.03, Ds[int(len(Ds)/2)], dx)

    p = wD.Vwave_mut_rate_pars(id_, tot_time=time, dt=dt, dx=dx, 
                     n_x_bins=int( ut.lin_from_two_points(r0, 0.05, 10, 1, 100)/dx ), 
                     M=5, r0=r0, alpha=0, beta=beta, gamma=gamma, lambda_tilde=lambda_tilde,
                     eps=eps, Ds=Ds, D0_i=D0_i,
                     Nh=10**12, N0=10**10, cutoff=cutoff,
                     t_burn=ut.lin_from_two_points(r0, 0.05, 250, 5, 5000), 
                     t_burn_cutoff=ut.lin_from_two_points(r0, 0.05, 70, 5, 1000), 
                    #t_burn=200, t_burn_cutoff=100,
                     back_width_fract=ut.lin_from_two_points(np.log(r0), np.log(0.05), 0.5, np.log(5), 3),
                     traj_step=int(50/dt), check_step=int(0/dt), traj_after_burn=False, verbose=True, global_cutoff=True)
    
    return deepcopy(wD.Vwave_mut_rate(p, True, init_n=init_n, init_nh=init_nh))

class eps_schedule():
    
    def __init__(self, first_eps, first_time, decay_exp, decay_scale):
        self.first_eps = first_eps
        self.first_time = first_time
        self.decay_exp = decay_exp
        self.decay_scale = decay_scale
        
    def get(self, t):
        if t < self.first_time:
            return self.first_eps
        else:
            t1 = t - self.first_time
            return self.first_eps*( self.decay_scale/(t1 + self.decay_scale) )**self.decay_exp
```

```python
def estf(d):
    k = -2.3381 * (d*s**2)**(1/3)
    return f*(1 - 2*l*d) + 2*k*(2/3 - l*d) - (a+g)*l*d
```

### Running the first series of stochastic simulations starting from a given initial condition

After this series, the other serires will use the final points of those one to save time for convergence.

```python
r0_list = np.logspace(np.log10(0.1), np.log10(7.5), 15)
lambda_tilde = 1000
n_D_bins = 150
time = 400000
eps = eps_schedule(2e-14, time/4, 1.5, time/4)
Ds = np.linspace(0.5e-4, 3e-4, n_D_bins)
```

```python
assays = []
for i, r0 in enumerate(r0_list):
    #Ds = np.linspace(0.8e-4, 2e-4, 60)
    assays.append(build_assay_wave(i, r0, 2, lambda_tilde, Ds, eps=eps, time=time))
```

```python
fig, ax = plt.subplots(1,1)
ax.set_xlabel('time')
ax.set_ylabel('mean D')

mean_D = []
av_time_start = 200000
speeds, f_tips, s_tips = np.array([]),np.array([]), np.array([])
est = np.array([])

for w in waves1:
    av_time_start_i = (w.traj.times < av_time_start).sum()
    speeds = np.append(speeds, np.mean(w.traj.speed(100)[av_time_start_i:]))
    f_tips = np.append(f_tips, np.mean(w.traj.f_tip[av_time_start_i:]))
    s_tips = np.append(s_tips, np.mean(w.traj.s_tip[av_time_start_i:]))
    a, g, l, b = w.p.alpha, w.p.gamma, w.p.lambda_tilde, w.p.beta
    v, f, s = speeds[-1], f_tips[-1], s_tips[-1]

    est =  np.append(est, brentq(estf, 0.5e-4, 3e-4))
    mean_D.append(np.median(w.traj.mean_D[av_time_start_i:]))
    ax.plot(w.traj.times - w.p.t_burn, w.traj.mean_D)
```

## Integrating all the runs

```python
r0_list = np.logspace(np.log10(0.1), np.log10(7.5), 15)
```

```python
f = open('data/mut_rate_stoch_assays.pickle', 'rb')
waves1 = pickle.load(f)
f.close()
f = open('data/mut_rate_stoch_assays2.pickle', 'rb')
waves2 = pickle.load(f)
f.close()
f = open('data/mut_rate_stoch_assays3.pickle', 'rb')
waves3 = pickle.load(f)
f.close()
f = open('data/mut_rate_stoch_assays4.pickle', 'rb')
waves4 = pickle.load(f)
f.close()

waves_all = [waves1, waves2, waves3, waves4]
av_time_starts = [200000, 100000, 100000, 100000]
```

```python
mean_D, std_D = np.zeros(len(r0_list)), np.zeros(len(r0_list))
speeds, f_tips, s_tips = np.zeros(len(r0_list)),np.zeros(len(r0_list)), np.zeros(len(r0_list))
est = np.zeros(len(r0_list))

tot_time = 0
for i, waves in enumerate(waves_all):
    av_time_start_i = (waves[0].traj.times < av_time_starts[i]).sum()
    time_average = float(waves[0].traj.times[-1] - waves[0].traj.times[av_time_start_i])
    tot_time += time_average
    
for i, waves in enumerate(waves_all):
    
    av_time_start_i = (waves[0].traj.times < av_time_starts[i]).sum()
    time_average = float(waves[0].traj.times[-1] - waves[0].traj.times[av_time_start_i])
    
    for r0i, w in enumerate(waves):
        
        v = np.mean(w.traj.speed(100)[av_time_start_i:])
        f = np.mean(w.traj.f_tip[av_time_start_i:])
        s = np.mean(w.traj.s_tip[av_time_start_i:])
        
        speeds[r0i] += v * time_average / tot_time
        f_tips[r0i] += f * time_average / tot_time
        s_tips[r0i] += s * time_average / tot_time
        
        a, g, l, b = w.p.alpha, w.p.gamma, w.p.lambda_tilde, w.p.beta
        est[r0i] +=  brentq(estf, 0.5e-4, 4e-4) * time_average / tot_time
        
        mean_D[r0i] += np.mean(w.traj.mean_D[av_time_start_i:]) * time_average / tot_time
        std_D[r0i] += np.std(w.traj.mean_D[av_time_start_i:]) * time_average / tot_time
```

```python
plt.xscale('log')
#plt.scatter(r0_list, mean_D, c='k', label='average D')
plt.errorbar(r0_list, mean_D, yerr=std_D)
plt.plot(r0_list, est)
plt.xlabel('r0')
plt.ylabel(r'$D_{ES}$')
```

```python
f = open('data/ES_mut_rate_stoch.tsv', 'w')

header = '#r0\tD_ES\tD_ES_std\tD_ES_tip\tF_T\ts_T\n'
f.write(header)

for i in range(len(r0_list)):
    f.write(str(r0_list[i]) + '\t')
    f.write(str(mean_D[i]) + '\t')
    f.write(str(std_D[i]) + '\t')
    f.write(str(est[i]) + '\t')
    f.write(str(f_tips[i]) + '\t')
    f.write(str(s_tips[i]) + '\t')
        
    f.write('\n')
    
f.close()
```
